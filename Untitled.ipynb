```python
import copy

def calculate_national_insurance(deductable_income, monthly) -> float:
    total_deductions = 0.0

    low_rate = 0.02
    high_rate = 0.08
    if monthly:
        high_rate_limit = 4_189.00
    else:
        high_rate_limit = 50_270.00

    if deductable_income > high_rate_limit:
        total_deductions += (deductable_income - high_rate_limit) * low_rate
        deductable_income = high_rate_limit

    total_deductions += deductable_income * high_rate

    return total_deductions


def calculate_income_tax_deductions(deductable_income, monthly) -> tuple[float, list]:
    """
        simplifying again here to include only the calculation for my tax code (the default, 1275L).
        using tax year 2025/26 values.
    """
    total_deductions = []
    
    basic_rate = 0.20
    higher_rate = 0.40
    additional_rate = 0.45
    if monthly:
        higher_rate_lower = 3_141.00
        higher_rate_upper = 10_428.00
    else:
        higher_rate_lower = 37_700.00
        higher_rate_upper = 125_140.00

    if deductable_income > higher_rate_upper:
        # charge additional rate on this portion
        additional_rate_deduction = (deductable_income - higher_rate_upper) * additional_rate
        total_deductions.append(additional_rate_deduction)
        deductable_income = higher_rate_upper
    else:
        total_deductions.append(0.00)

    if deductable_income > higher_rate_lower:
        # charge higher rate on this portion
        high_rate_deduction = (deductable_income - higher_rate_lower) * higher_rate
        total_deductions.append(high_rate_deduction)
        deductable_income = higher_rate_lower
    else:
        total_deductions.append(0.00)

    # charge basic rate on the rest
    basic_rate_deduction = deductable_income * basic_rate
    total_deductions.append(basic_rate_deduction)

    return sum(total_deductions), total_deductions


def calculate_student_loan_repayment(gross_salary, monthly) -> float:
    """
        simplifying here by only including the [Plan](https://www.gov.uk/repaying-your-student-loan/when-you-start-repaying)
        relevant to me (i.e. Plan 1), and also assuming i'll be earning enough to trigger repayment (salary > £26K) 
        result: 9% (https://www.gov.uk/repaying-your-student-loan/what-you-pay).
        note: the 
    """
    rate = 0.09
    if monthly:
        allowance = 2_172.00
    else:
        allowance = 26_065.00
    deductable_salary = gross_salary - allowance
    
    return deductable_salary * rate


def calculate_net_salary(gross_salary, pension_contrib_prct=0.05, monthly=False, verbose=False) -> float:
    """
        flow:
            gross
            |___ Other
                |___ Pension
                |___ Student Loan
                |___ Personal Allowance
            |___ Income
                |___ National Insurance
                |___ Brackets
                    |___ 20%
                    |___ 40%
                    |___ 45%
    """
    total_deductions_other = 0.0
    pension_contrib = gross_salary * pension_contrib_prct
    student_loan_repayment = calculate_student_loan_repayment(gross_salary, monthly)
    if monthly:
        personal_allowance = 1_062.50
    else:
        personal_allowance = 12_750.00
    total_deductions_other += pension_contrib
    total_deductions_other += student_loan_repayment
    
    deductable_salary = gross_salary - personal_allowance - pension_contrib

    total_deductions_income = 0.0
    national_insurance = calculate_national_insurance(deductable_salary, monthly)
    income_tax_deductions_total, income_tax_deductions_parts = calculate_income_tax_deductions(deductable_salary, monthly)
    total_deductions_income += income_tax_deductions_total
    total_deductions_income += national_insurance

    total_deductions = total_deductions_other + total_deductions_income
    net_salary = gross_salary - total_deductions

    if verbose:
        print(f"gross: {gross_salary:,.2f} ({total_deductions:,.2f})")
        print(f"|___ {(total_deductions_other):,.2f} ")
        print(f"    |___ Pension: {pension_contrib:,.2f}")
        print(f"    |___ Student Loan: {student_loan_repayment:,.2f}")
        print(f"    |___ Personal Allowance: {personal_allowance:,.2f}")
        print(f"|___ {deductable_salary:,.2f} ({(total_deductions_income):,.2f})")
        print(f"    |___ National Insurance: {national_insurance:,.2f}")
        print(f"    |___ Brackets: {income_tax_deductions_total:,.2f}")
        print(f"        |___ tax at 20%: {income_tax_deductions_parts[-1]:,.2f}")
        print(f"        |___ tax at 40%: {income_tax_deductions_parts[-2]:,.2f}")
        print(f"        |___ tax at 45%: {income_tax_deductions_parts[-3]:,.2f}")
        print(f"-----------------------------------------------------")
        print(f"net: {net_salary:,.2f}")

    return net_salary
    

def recommend_budget(income):
    """
    gonna go for the 50:30:20
    
        income
        |___ Essentials
            |___ Rent
            |___ Bills
            |___ Groceries
            |___ Phone
            |___ Medical Bills
            |___ Public Transport
            |___ Gym
        |___ Savings
            |___ Savings
            |___ Investments
        |___ Spending
            |___ Eating Out
            |___ Clothes
            |___ Flowers
            |___ Pub
    """
    cost_ratios = {
        "Essentials": {
            "Rent": 0.35,
            "Bills": 0.05,
            "Groceries": 0.07,
            "Phone": 0.01,
            "Medical Bills": 0.01,
            "Gym": 0.01
        },
        "Savings": {
            "Savings": 0.1,
            "Investments": 0.1
        },
        "Spending": {
            "Eating Out": 0.05,
            "Pub": 0.05,
            "Flowers": 0.05,
            "Clothes": 0.05,
            "Skincare": 0.05,
            "Vitamins": 0.05
        }
    }

    costs = copy.deepcopy(cost_ratios)
    budgeted = 0.0

    for category, items in costs.items():
        for entry, value in items.items():
            costs[category][entry] = income * value

    print(f"{income:,.2f}")
    for category, items in costs.items():
        print(f"|___ {category}: {sum(items.values()):,.2f} ({(sum(items.values())/income)*100:,.2f}%)")
        for entry, value in items.items():
            print(f"    |___ {entry}:\t{value:,.2f}\t({(value/income)*100:,.2f}%)")
            budgeted += value
    
    leftover = income - budgeted
    print(f"leftover: {leftover:,.2f}")

    return costs


def create_budget(income):
    """
        income
        |___ Essentials
            |___ Rent
            |___ Bills
            |___ Groceries
            |___ Phone
            |___ Medical Bills
            |___ Public Transport
            |___ Gym
        |___ Savings
            |___ Savings
            |___ Investments
        |___ Spending
            |___ Eating Out
            |___ Clothes
            |___ Flowers
            |___ Pub
    """
    costs = {
        "Essentials": {
            "Rent": 850.00,
            "Bills": 100.00,
            "Groceries": 500.00,
            "Phone": 30.0,
            "Medical Bills": 30.0,
            "Gym": 30.0
        },
        "Savings": {
            "Savings": 0.00,
            "Investments": 0.00
        },
        "Spending": {
            "Eating Out": 250.00,
            "Pub": 100.00,
            "Flowers": 250.00,
            "Clothes": 150.00,
            "Skincare": 150.00,
            "Vitamins": 100.00
        }
    }
    budgeted = 0.0

    print(f"{income:,.2f}")
    for category, items in costs.items():
        print(f"|___ {category}: {sum(items.values()):,.2f} ({(sum(items.values())/income)*100:,.2f}%)")
        for entry, value in items.items():
            print(f"    |___ {entry}:\t{value:,.2f}\t({(value/income)*100:,.2f}%)")
            budgeted += value
    
    leftover = income - budgeted
    print(f"leftover: {leftover:,.2f}")


def calculate_national_insurance(deductable_income, monthly) -> float:
    total_deductions = 0.0

    low_rate = 0.02
    high_rate = 0.08
    if monthly:
        high_rate_limit = 4_189.00
    else:
        high_rate_limit = 50_270.00

    if deductable_income > high_rate_limit:
        total_deductions += (deductable_income - high_rate_limit) * low_rate
        deductable_income = high_rate_limit

    total_deductions += deductable_income * high_rate

    return total_deductions


def calculate_income_tax_deductions(deductable_income, monthly) -> tuple[float, list]:
    """
        simplifying again here to include only the calculation for my tax code (the default, 1275L).
        using tax year 2025/26 values.
    """
    total_deductions = []
    
    basic_rate = 0.20
    higher_rate = 0.40
    additional_rate = 0.45
    if monthly:
        higher_rate_lower = 3_141.00
        higher_rate_upper = 10_428.00
    else:
        higher_rate_lower = 37_700.00
        higher_rate_upper = 125_140.00

    if deductable_income > higher_rate_upper:
        # charge additional rate on this portion
        additional_rate_deduction = (deductable_income - higher_rate_upper) * additional_rate
        total_deductions.append(additional_rate_deduction)
        deductable_income = higher_rate_upper
    else:
        total_deductions.append(0.00)

    if deductable_income > higher_rate_lower:
        # charge higher rate on this portion
        high_rate_deduction = (deductable_income - higher_rate_lower) * higher_rate
        total_deductions.append(high_rate_deduction)
        deductable_income = higher_rate_lower
    else:
        total_deductions.append(0.00)

    # charge basic rate on the rest
    basic_rate_deduction = deductable_income * basic_rate
    total_deductions.append(basic_rate_deduction)

    return sum(total_deductions), total_deductions


def calculate_student_loan_repayment(gross_salary, monthly) -> float:
    """
        simplifying here by only including the [Plan](https://www.gov.uk/repaying-your-student-loan/when-you-start-repaying)
        relevant to me (i.e. Plan 1), and also assuming i'll be earning enough to trigger repayment (salary > £26K) 
        result: 9% (https://www.gov.uk/repaying-your-student-loan/what-you-pay).
        note: the 
    """
    rate = 0.09
    if monthly:
        allowance = 2_172.00
    else:
        allowance = 26_065.00
    deductable_salary = gross_salary - allowance
    
    return deductable_salary * rate


def calculate_net_salary(gross_salary, pension_contrib_prct=0.05, monthly=False, verbose=False) -> float:
    """
        flow:
            gross
            |___ Other
                |___ Pension
                |___ Student Loan
                |___ Personal Allowance
            |___ Income
                |___ National Insurance
                |___ Brackets
                    |___ 20%
                    |___ 40%
                    |___ 45%
    """
    total_deductions_other = 0.0
    pension_contrib = gross_salary * pension_contrib_prct
    student_loan_repayment = calculate_student_loan_repayment(gross_salary, monthly)
    if monthly:
        personal_allowance = 1_062.50
    else:
        personal_allowance = 12_750.00
    total_deductions_other += pension_contrib
    total_deductions_other += student_loan_repayment
    
    deductable_salary = gross_salary - personal_allowance - pension_contrib

    total_deductions_income = 0.0
    national_insurance = calculate_national_insurance(deductable_salary, monthly)
    income_tax_deductions_total, income_tax_deductions_parts = calculate_income_tax_deductions(deductable_salary, monthly)
    total_deductions_income += income_tax_deductions_total
    total_deductions_income += national_insurance

    total_deductions = total_deductions_other + total_deductions_income
    net_salary = gross_salary - total_deductions

    if verbose:
        print(f"gross: {gross_salary:,.2f} ({total_deductions:,.2f})")
        print(f"|___ {(total_deductions_other):,.2f} ")
        print(f"    |___ Pension: {pension_contrib:,.2f}")
        print(f"    |___ Student Loan: {student_loan_repayment:,.2f}")
        print(f"    |___ Personal Allowance: {personal_allowance:,.2f}")
        print(f"|___ {deductable_salary:,.2f} ({(total_deductions_income):,.2f})")
        print(f"    |___ National Insurance: {national_insurance:,.2f}")
        print(f"    |___ Brackets: {income_tax_deductions_total:,.2f}")
        print(f"        |___ tax at 20%: {income_tax_deductions_parts[-1]:,.2f}")
        print(f"        |___ tax at 40%: {income_tax_deductions_parts[-2]:,.2f}")
        print(f"        |___ tax at 45%: {income_tax_deductions_parts[-3]:,.2f}")
        print(f"-----------------------------------------------------")
        print(f"net: {net_salary:,.2f}")

    return net_salary


def main(salary):
    yearly_salary = calculate_net_salary(salary)
    monthly_salary = calculate_net_salary(salary/12, monthly=True)
    print(f"salary:  {salary:,.2f}")
    print(f"yearly:  {yearly_salary:,.2f}")
    print(f"monthly: {monthly_salary:,.2f}")
    print()
    print("current budget:")
    create_budget(monthly_salary)
    print()
    print("recommended budget:")
    recommend_budget(monthly_salary)
    return monthly_salary
```
